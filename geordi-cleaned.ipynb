### Calculating levels in production classification csv

```python
workflow_names=production_classification.workflow_name.unique()# stores the classification workflow names
```

```python
workflow_names
```

```python
levels=[]# Calculating levels corresponding to the workflow name
```

Here from level we equate the workflow to the level of classification the user is doing currently

```python
for xname in production_classification.workflow_name:
  if xname==workflow_names[0]:
    levels.append(1)
  elif xname==workflow_names[1]:
    levels.append(2)
  elif xname==workflow_names[2]:
    levels.append(3)
  elif xname==workflow_names[3] or xname==workflow_names[7]:
    levels.append(4)
  elif xname==workflow_names[4] or xname==workflow_names[9]:
    levels.append(5)
  elif xname==workflow_names[8]:
    levels.append(6)
  elif xname==workflow_names[10]:
    levels.append(7)
  elif xname == workflow_names[5]:
    levels.append(0)
  elif xname == workflow_names[6]:
    levels.append(0)
```

```python
production_classification['levels'] = levels
```
